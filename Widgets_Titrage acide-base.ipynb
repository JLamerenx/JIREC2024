```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets


#Données
#=========
pKe = 14          # Produit ionique de l'eau
pKa1 = 1.2        # Constante d'acidité
pKa2 = 4.3

Ke  = 10**(-pKe)
Ka1 = 10**(-pKa1)
Ka2 = 10**(-pKa2)

ca = 0.1          # Concentration de l'acide en mol/L
Va = 10           # Volume titré en mL
cb = 0.1          # Concentration de la soude en mol/L


#Simulation de la courbe
#========================
pH_sim = np.linspace(0,13,500)
h = 10**-pH_sim
numerateur = ca*Va * (2+h/Ka2) / (1+h/Ka2+h**2/(Ka1*Ka2)) + Va * (Ke/h-h)
denominateur = cb + h - Ke/h
Vb = numerateur / denominateur


#Valeurs expérimentales
#=======================
V = np.array([0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.2,8.4,8.6,8.8,9,9.2,9.4,9.6,9.8,10,10.2,10.4,10.6,10.8,11,11.2,11.4,11.6,11.8,12,12.5,13,13.5,14,14.5,15,15.5,16,16.5,17,17.2,17.4,17.6,17.8,18,18.2,18.4,18.6,18.8,19,19.2,19.4,19.6,19.8,20,20.2,20.4,20.6,20.8,21,21.2,21.4,21.6,21.8,22,22.5,23,23.5,24,24.5,25])
pH = np.array([1.35,1.39,1.41,1.45,1.48,1.51,1.55,1.59,1.63,1.68,1.72,1.77,1.83,1.9,1.96,2.06,2.14,2.19,2.23,2.28,2.33,2.38,2.44,2.5,2.57,2.61,2.68,2.76,2.81,2.87,2.93,2.97,3.05,3.1,3.15,3.2,3.25,3.36,3.45,3.55,3.63,3.73,3.81,3.91,4,4.1,4.21,4.27,4.3,4.37,4.42,4.46,4.56,4.62,4.71,4.8,4.89,5.04,5.36,5.7,6.78,9.97,10.62,10.95,11.17,11.31,11.41,11.47,11.55,11.62,11.67,11.71,11.81,11.89,11.95,12.01,12.05,12.09])


#Calcul des proportions des espèces acido-basiques exprimées en %
#=================================================================
fAH2 = 100 / (1 + 10**(pH-pKa1) + 10**(2*pH-pKa1-pKa2))
fAH = 100 / (1 + 10**(pKa1-pH) + 10**(pH-pKa2))
fA = 100 - fAH2 - fAH


#Génération du tracé
#====================

def courbe_avec_simul(distribution,position,simulation) :

  k = position

  fig, ax1 = plt.subplots()
  ax2 = ax1.twinx()

  ax1.plot(V,pH,'b+',label='$pH$')
  ax1.set_xlabel('Volume V (mL)')
  ax1.set_ylabel('pH')
  ax1.plot(V[k],pH[k],'bo')
  ax1.set_xlim(0,25)
  ax1.set_ylim(0,13)
  ax1.grid(True)

  if distribution == 'Oui' :
    ax2.plot(V,fAH2,'g--',label='$AH_2$')
    ax2.plot(V[k],fAH2[k],'go')
    ax2.plot(V,fAH,'r--',label='$AH$')
    ax2.plot(V[k],fAH[k],'ro')
    ax2.plot(V,fA,'y--',label='$A$')
    ax2.plot(V[k],fA[k],'yo')
    ax2.legend()
    ax2.set_ylabel('Proportions')

  if simulation == 'Oui' :
    ax1.plot(Vb,pH_sim,'+',ms=1,label='simulation')

  plt.title(f"Suivi pH-métrique du titrage d'une solution d'acide oxalique \n Valeurs des $pK_a$ = {pKa1} et {pKa2}")
  plt.show()


bouton = widgets.RadioButtons(options=['Oui', 'Non'], value = 'Non',description='Distribution')
simulation = widgets.RadioButtons(options=['Oui', 'Non'], value = 'Non',description='Simulation')
position = widgets.IntText(value=0,min=0,max=len(V),step=2,description='Position')


out = widgets.interactive_output(courbe_avec_simul, {'distribution': bouton,'position':position,'simulation':simulation})
visuel = widgets.HBox([widgets.VBox([bouton,simulation,position]), out])
visuel

```
